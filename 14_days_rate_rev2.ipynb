```python
from scipy import stats
import matplotlib.pyplot as plt

pd.options.mode.chained_assignment = None  # default='warn' -> to fix SettingwithCopyWarning error

# Create two datasets from two urls
df = pd.read_csv(url_covid_general)

df2 = pd.read_csv(url_14days)
df2.sort_index(ascending=False, inplace=True)
```

```python
def model_regression(country,days_number=100):
    '''
    Function returns linear and polynomial regression for selected country
    and prediction of test's positive rate a number of days future
    '''
    # country name saved separatly to use as a string in the plots's title
    country_name = country
    
    # Create dataframe with dataset of selected country
    country1 = df[df['location'] == country]

    # Create series of date and convert the dates to 0-based number of days from 31st December 2019
    country_date = country1['date']
    country_days = [x for x in range(0,len(country_date))]

    # Repleace all nan by 0.00
    country_positive_rate = country1['positive_rate'].fillna(0)
    
    ####################################################
    
    # Create dataframe with dataset of selected country
    country2 = df2[df2['countriesAndTerritories'] == country]

    # Create series of date and convert the dates to 0-based number of days from 31st December 2019
    country_date2 = country2['dateRep']
    country_days2 = [x for x in range(0,len(country_date2))]

    # Repleace all nan by 0.00
    country_14days_per_100k = country2['notification_rate_per_100000_population_14-days'].fillna(0)
    
    # rename column in second dataframe
    country2.rename(columns={'dateRep': 'date'},inplace=True)

    # convert columne with date (strings) to datetime object
    country1.date = pd.to_datetime(country1.date)
    country2.date = pd.to_datetime(country2.date,dayfirst=True) # dayfirst parameter set to True fixed the issue with day and month which were mixing
    
    # merge data frames on date column
    merged_dfs = pd.merge(country1, country2, how='inner', on='date')
    #return merged_dfs
       
    ###################################################################
    #############         LINEAR REGRESSION       #####################
    ###################################################################

    %matplotlib inline

    from IPython.display import set_matplotlib_formats
    set_matplotlib_formats('svg')
    from scipy import stats
    
    ########################################
    location_date = merged_dfs['date']
    location_days = [x for x in range(0,len(location_date))]
    location_positive_rate = merged_dfs['positive_rate'].fillna(0) # repleace all nan with '0' -> if not change linear_reg_positive_rate list will be empty
    slope, intercept, r, p, se = stats.linregress(location_days, location_positive_rate)

    def f(x):
        return slope * x + intercept

    linear_reg_positive_rate = list(map(f, location_days))
    
    ##########################################
    location_14days_per_100k = merged_dfs['notification_rate_per_100000_population_14-days']
    slope2, intercept2, r2, p2, se2 = stats.linregress(location_days, location_14days_per_100k)

    def f2(x):
        return slope2 * x + intercept2

    linear_reg_14days = list(map(f2, location_days))
    
    ###################################################################
    #############       POLYNOMIAL REGRESSION       ###################
    ###################################################################
    
    # Do a polynomial regression analysis and plot the polynomial line on the same plot

    # Create a function which calculate y values
    import numpy as np
    mymodel_test_rate = np.poly1d(np.polyfit(location_days,location_positive_rate,3))

    # Create 100 x values in a range of weigt values
    x_values = np.linspace(min(location_days),max(location_days),100)

    # Calculate y values
    y_values = mymodel_test_rate(x_values)
    y_values
    
    mymodel_14days = np.poly1d(np.polyfit(location_days,location_14days_per_100k,3))

    # Create 100 x values in a range of weigt values
    x_14days = np.linspace(min(location_days),max(location_days),100)

    # Calculate y values
    y_14days = mymodel_14days(x_14days)
    y_14days
    
    ###################################################################
    #############          FORMATTING             #####################
    ###################################################################
    
    plt.style.use('ggplot')
    #plt.figure(figsize=(15,10), facecolor='grey')
    fig, axs = plt.subplots(nrows=2, sharex=True)
    
    fig.set(facecolor='grey')
    fig.suptitle(f'Regression model for {country_name}',color='white',fontsize=20, va='bottom')
    
    axs[0].plot(location_days,location_positive_rate)
    axs[0].plot(location_days, linear_reg_positive_rate)
    axs[0].plot(x_values,y_values)
    axs[0].set_title('Positive Test\'s Rate')
    axs[0].legend(['positive test\'s rate','linear regression','polynomial regression'])
    axs[1].plot(location_days,location_14days_per_100k)
    axs[1].plot(location_days, linear_reg_14days)
    axs[1].plot(x_14days,y_14days)
    axs[1].set_title('14 Days Notification Rate ')
    
    # Show the r-squared value
    r_sq_l_positive = r**2
    r_sq2_l_14days = r2**2
    
    # Show the polynomial r-squared value
    from sklearn.metrics import r2_score
    
    r_sq_p_positive = r2_score(location_positive_rate,mymodel_test_rate(location_days))
    r_sq_p_14days = r2_score(location_14days_per_100k,mymodel_14days(location_days))
    
    ###################################################################
    #############             PREDICTIONS         #####################
    ###################################################################
    
    # Prediction of total cases from today
    import datetime
    today = datetime.datetime.today()
    today_date = datetime.date.today()
    first_day = country_date.iloc[0]
    first_day = datetime.datetime.strptime(first_day, '%Y-%m-%d')
    delta = (today - first_day).days
    future_date = today_date + datetime.timedelta(days=days_number)
    
    # define function to calculate what would be a traffic light for the country based on positive rate of daily tests
    # and 14 day notification rate (cumulative number of COVID-19 cases per 100,000 population)
    
    def traffic_light(no_cases_per_100k, positive_test_rate):
        '''Return ‘traffic light’ for the selected country in the future based on positive rate of daily tests and 14 day notification rate'''
        
        no_cases_per_100k = round(no_cases_per_100k,0)
        positive_test_rate = round(positive_test_rate,4)*100
        if (no_cases_per_100k < 25) and (positive_test_rate < 4):
            print(f'with {no_cases_per_100k} cases and {positive_test_rate}% of positive rate {country_name} is on GREEN list')
        
        elif ((no_cases_per_100k < 50) and (positive_test_rate >= 4)) or ((no_cases_per_100k < 150) and (positive_test_rate < 4)):
            print(f'with {no_cases_per_100k} cases and {positive_test_rate}% of positive rate {country_name} is on ORANGE list')
        
        elif ((no_cases_per_100k >= 50) and (positive_test_rate) >=4) or (no_cases_per_100k > 150):
            print(f'with {no_cases_per_100k} cases and {positive_test_rate}% of positive rate {country_name} is on RED list')
        
        else:
            print(f'with this rates your country is on GREY list')
    
    positive_test_rate_lin = float(f(days_number+delta))
    no_cases_per_100k_lin = float(f2(days_number+delta))
    
    positive_test_rate_pol = float(mymodel_test_rate(days_number+delta))
    no_cases_per_100k_pol = float(mymodel_14days(days_number+delta))
     
    print(f'COUNTRY NAME: {country_name}')
    print(f'DATE: {today_date}')
    print(f'FUTURE DATE: {future_date}')
    print(f'NUMBER OF DAYS IN FUTURE: {days_number}')
    
    print('\n**********LINEAR REGRESSION**********\n')
    print(f'R-squared (postive rate): {r_sq_l_positive:.2}')
    print(f'R-squared (14days rate): {r_sq2_l_14days:.2}')
    print(f'Positive Test Rate: {positive_test_rate_lin:.2}')
    print(f'14 days notification: {round(no_cases_per_100k_lin,2)}\n')        
    print('TRAFFIC LIGHT:')
    print(traffic_light(no_cases_per_100k_lin,positive_test_rate_lin))
    
    print('\n**********POLYNOMIAL REGRESSION:**********\n')
    print(f'R-squared (postive rate): {r_sq_p_positive:.2}')
    print(f'R-squared (14days rate): {r_sq_p_14days:.2}')
    print(f'Positive Test Rate: {positive_test_rate_pol:.2}')
    print(f'14 days notification: {round(no_cases_per_100k_pol,2)}\n')
    print('TRAFFIC LIGHT')
    print(traffic_light(no_cases_per_100k_pol,positive_test_rate_pol))
    print("")
    last_day_tests = merged_dfs['positive_rate'].iloc[-1]
    last_day_14days = merged_dfs['notification_rate_per_100000_population_14-days'].iloc[-1]
    print(f'*****TRAFFIC LIGHT on {today_date} for {country_name}*****')
    return traffic_light(last_day_14days,last_day_tests)
    
    
    # Calculate what would be a ‘traffic light’ for the selected country in the future based on positive rate of daily tests 
    # and 14 day notification rate (cumulative number of COVID-19 cases per 100,000 population.
```
